```python
def area(no_side=0):
    import math
    if no_side == 1:
        radius = int(input("enter a radius"))
        print(f"area of circle is   {3.14*radius**2}")
    elif no_side == 3:
        base = int(input("enter a base "))
        height = int(input("enter a height "))
        print(f"area of triangle is   {0.5*base*heght}")
    elif no_side == 4:
        length = int(input("enter a length"))
        breath = int(input("enter a breath"))
        if length == breath:
            print(f"area of square is  {length*breath}")
        else:
            print(f"area of ractangle is  {length*breath}")
    elif no_side == 5:
        side = int(input("enter a side"))
        print(f" area of pentagone is  {(1/4) * math.sqrt(5 * (5 + 2 * math.sqrt(5))) * side**2}")
    elif no_side == 6:
        side = int(input("enter a side "))
        print(f"area of hexagone is  {(3 * math.sqrt(3) * side**2) / 2}")
    elif no_side == 8:
        side = int(input("enter a side"))
        area = area = 2 * (1 + (2 ** 0.5)) * side**2         
        return f"area of octagone is {area}"
```

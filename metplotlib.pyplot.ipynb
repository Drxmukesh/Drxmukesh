```python
x=[i for i in range(10)]
y=[i**2 for i in range(10)]
z=[i*5 for i  in range(10)]

plt.figure(figsize=[12,5])
plt.plot(x,y,label="square",marker=5)
plt.plot(x,z,label="cube",marker=5)

plt.title("line plot \n",fontdict={"fontsize":24,"fontweight":"bold"})

plt.xlabel("values")
plt.ylabel("square")

plt.legend()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([8,6,14,4,12])
label1=["deva","roma","koma","kelu","gela"]
myexplode=[0.2,0.1,0.01,0.001,0.0001]
col=["#8c3861","#79b791","#2f4b26","#dbf4a1","#dbf4a4"]
plt.pie(y,labels=label1,explode=myexplode,colors=col)
plt.show() 
```

```python
print(df1["Embarked"].value_counts(normalize=True))
plt.pie(df1["Embarked"].value_counts(),
       labels=df1["Embarked"].unique()[:-1],
       autopct="%1.2f%%",
       startangle=90,
       explode=[0.01,0.1,0.02]
       )
plt.legend()
```

```python
import pandas as pd
left = pd.DataFrame(
    {
        "left_id":[10,20,30,40,50],
        "l_name":["dog","cat","rat","fox","cow"],
        "sid":["s1","s2","s4","s6","s5"]
    }
)
left
```

```python
right = pd.DataFrame(
    {
        "right_id":[60,70,80,90,5],
        "r_name":["dog","cat","bat","hen","cow"],
        "sid":["s7","s4","s3","s6","s5"]
    }
)
right
```

```python
pd.merge(left  ,right,on="sid",how="inner")
```

```python
pd.merge(left,right,on="sid",how="left")
```

```python
pd.merge(left,right,on="sid",how="right")
```

```python
pd.merge(left,right,on="sid",how="outer")
```

```python
import numpy as np
x = np.arange(0, 200)
y = np.arange(0, 200) 

fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])       
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2]) 

ax1.plot(x, y, color='red', label='Main Axis')
ax1.set_title('Main Axis')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')

ax2.plot(x, y, color='red', label='Small Axis')
ax2.set_title('Small Axis')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')

plt.show()
```

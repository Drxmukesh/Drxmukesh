```python
plt.figure(figsize=(15,7))
sns.countplot(x='season',
              data=matches,
              order = matches['season'].value_counts().index.sort_values()
              )
plt.xlabel('season',fontsize=12)
plt.ylabel('no. of matches',fontsize=12)
plt.title('no. of matches played in each IPL season',fontsize=16)
```

```python
plt.figure(figsize = (11,9))
sns.countplot(y = 'winner',
              data = matches,
              order= matches['winner'].value_counts().index,
              palette = 'copper' )
plt.xlabel('no. Wins',fontsize=12)
plt.ylabel('team',fontsize=12)
plt.title('matches won teams',fontsize=16)
```

```python
plt.figure(figsize = (12,8))
sns.countplot(y = 'venue',
              data = matches,
              order = matches['venue'].value_counts().iloc[:10].index,
              palette='gist_heat')
plt.xlabel('no. of matches',fontsize=12)
plt.ylabel('venue',fontsize=12)
plt.title('no. of matches played on each Stadium',fontsize=16)
```

```python
plt.figure(figsize=(6,6))
sizes = matches.toss_decision.value_counts()
labels = matches.toss_decision.value_counts().index
plt.pie(sizes,colors = ['red','blue'],
         labels=labels,
         autopct='%1.1f%%',
         startangle=90,
         pctdistance=0.75,
         explode = (0.025,0.025))
centre_circle = plt.Circle((0,0),0.60,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
```

```python
plt.figure(figsize=(15,7))
sns.countplot(x='season',
              hue='toss_decision',
              data=matches,
              order =matches['season'].value_counts().index.sort_values(),
              palette='deep')
plt.xlabel('season',fontsize=12)
plt.ylabel('count',fontsize=12)
plt.title('decision to field or bat in each IPL season', fontsize=16)
```

```python
merge_data = pd.merge(matches,delivery, left_on='id', right_on='match_id', how='inner')
total_runs = merge_data[merge_data['batsman_runs'] == 4]['batsman_runs'].sum()
fours_hit = merge_data[merge_data['batsman_runs'] == 4].groupby('batting_team').size()

plt.figure(figsize=(10, 6))
fours_hit.plot(kind='bar', color='skyblue')
plt.title('total no. of hit 4 by each team')
plt.xlabel('team')
plt.ylabel('total Hit 4')
plt.xticks(rotation=60)
plt.grid(axis='y', linestyle='--', alpha=0.7)

for i, value in enumerate(fours_hit):
    plt.text(i, value + 1, str(value), ha='center', va='bottom')

plt.tight_layout()
plt.show()

print("total runs by 4:", total_runs)
print("total no. of hit 4 by each team:")
print(fours_hit)
```

```python
merge_data = pd.merge(matches,delivery, left_on='id', right_on='match_id', how='inner')
hit_player_4 = merge_data[merge_data['batsman_runs'] == 4].groupby('batsman').size()

plt.figure(figsize=(12, 6))
hit_player_4.sort_values(ascending=False)[:20].plot(kind='bar', color='skyblue')
plt.title('Total Fours Hit by Players')
plt.xlabel('Player')
plt.ylabel('Total Fours Hit')
plt.xticks(rotation=60)
plt.grid(axis='y', linestyle='--', alpha=0.7)
for i, value in enumerate(hit_player_4.sort_values(ascending=False)[:20]):
    plt.text(i, value + 1, str(value), ha='center', va='bottom')

plt.tight_layout()
plt.show()

print("Total number of fours hit by each player:")
hit_player_4
```

```python
merge_data = pd.merge(matches,delivery, left_on='id', right_on='match_id', how='inner')
fours_hit_per_season = merge_data[merge_data['batsman_runs'] == 4].groupby('season').size()
plt.figure(figsize=(10, 6))
fours_hit_per_season.plot(kind='bar', color='skyblue')
plt.title('Number of Fours Hit in Each Season')
plt.xlabel('Season')
plt.ylabel('Number of Fours Hit')
plt.grid(axis='y', linestyle='--', alpha=0.7)

for i, value in enumerate(fours_hit_per_season):
    plt.text(i, value + 1, str(value), ha='center', va='bottom')

plt.tight_layout()
plt.show()

print("Number of fours hit in each season:")
print(fours_hit_per_season)
```

```python
merged_data = pd.merge(matches, delivery, left_on='id', right_on='match_id', how='inner')
total_runs_from_sixes = merged_data[merged_data['batsman_runs'] == 6]['batsman_runs'].sum()

sixes_hit_by_team = merged_data[merged_data['batsman_runs'] == 6].groupby('batting_team').size()

plt.figure(figsize=(10, 6))
sixes_hit_by_team.plot(kind='bar', color='skyblue')
plt.title('Total Sixes Hit by Each Team')
plt.xlabel('Team')
plt.ylabel('Total Sixes Hit')
plt.xticks(rotation=30)
plt.grid(axis='y', linestyle='--', alpha=0.7)

for i, value in enumerate(sixes_hit_by_team):
    plt.text(i, value + 1, str(value), ha='center', va='bottom')

plt.tight_layout()
plt.show()

print("Total runs scored from sixes:", total_runs_from_sixes)
print("Total number of sixes hit by each team:")
print(sixes_hit_by_team)
```
